```python
import requests # requests module of python
import pandas as pd
import re
!pip install --upgrade pandas
import pandas as pd
import io
import numpy as np
```

```python
url = 'https://pastebin.com/V0SFsxRu' # url of paste

def rreplace(s, old, new): #To get the Url in RAW mode we replace the LAST / by /raw/
    return (s[::-1].replace(old[::-1],new[::-1], 1))[::-1]

raw_url = rreplace(url,'/','/raw/')
text = requests.get(raw_url).text # response will be stored as a TEXT
print(text) # prints content
```

```python
splited = text.split("PRIMARY KEY (`ApplicationID`)") #There we will split the text into 2 parts, the CREATE TABLE and the INSERT INTO
columns = splited[0]
values = splited[1].split("VALUES (")[1] # Here we take only the values
values = values.replace("\r\n", "") # Here we delete all the aleatory \r\n (End Of Line)
values = values.replace("),(", "\n") # Here we replace every end of row ),( for an \n in order to make a .csv format
print(values)
```

```python
import re
columns = re.findall("`\w+`", columns) #This will take all values under ` ` (quotes) and make a list
column_string = ','.join(columns[1:]) # We join the list maked before into a string separated by commas, and deleting the first one cause is the Table name
column_string = column_string.replace("`", "") #We dont need the ` ` (quotes) anymore
print(column_string)
```

```python
#Before continue I will check the format in order to clear the values and search if there is some errors.

# I detected that the prices in $ have the format $XX,XXX so in order to make it compatible with CSV will change commas to dots.
f = re.compile(r'(\d),(\d)')
clearvalues = f.sub(r'\1.\2', values) 

# I detected one single problem in the value Boston,MA with the comma so i excepcionally solve this by:
p = re.compile(r'(Boston), (MA)')
clearvalues = p.sub(r'\g<1>-\g<2>', clearvalues) 

clearvalues = clearvalues.replace(" ", "") #We delete all the spaces
clearvalues = clearvalues.replace("'", "") # We delete the quotes
clearvalues = clearvalues.replace(",,", ",") # We delete the double commas

#detected 1 values that makes problems
clearvalues = clearvalues.replace("Investments,Compensation", "Investments Compensation")

#For some reason there are 5 more values than columns in the Second Line, so we add extra columns to show what is happening.
string_dataframe = column_string + ",extra_columns,extra_columns,extra_columns,extra_columns,extra_columns" + "\n" + clearvalues
print(string_dataframe)
```

```python
#Here we create the dataframe Finally!!
data = io.StringIO(string_dataframe)
df = pd.read_csv(data, sep=",")
pd.set_option('display.max_columns', None) #This line makes to shows all the columns.
df
```

```python
# As we can see there are many useless columns full of chaos, mainly in the second line, where we can find for example the date 2/18/2014 repeated several times so we have to filter them:

#First we replace some values that means NaN but were not detected.
df = df.replace('None', np.NaN)
df = df.replace('NONE', np.NaN)
df = df.replace('NotSet', np.NaN)
df = df.replace(0.0 , np.NaN)

# Now we will delete all columns that have a percentage of NaN > 40% in order to make the final step, choose the desired columns to normalize.
df_clear = df.dropna(axis=1, thresh=int(0.4*len(df)))
df_clear
```

```python
#Columns to normalize:

#Columns not found to normalise: password_plain, password_hash, Fax , Secondary_last_name , date_update , date_miscellaneous
#Iban, Credit_card_number, Credit_card_last4  Credit_card_exp_date, credit_card_exp_month, Credit_card_exp_year, Credit_card_type , web_address, ip_ipv4 , ip_ipv6 , token , avatar
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Renaming columns to match the selected normalisation table:
df_trans = df_clear.rename(columns={'PrimaryFirstName':'First_name','PrimaryMiddleName':'Middle_name','PrimaryLastName':'Last_name','PrimaryEmail':'Email','UserName':'Username','PrimaryPhone_Home':'Phone','PrimaryGender':'Gender','PrimaryMaritalStatus':'Marital_status'})
df_trans = df_trans.rename(columns={'PrimaryDOB':'Date_of_birth','AppDate':'Date_register','PrimaryEmployerName':'Company','PrimaryEmployerStreet1':'Company_address','PrimaryOccupation':'Professional_role'})
df_trans = df_trans.rename(columns={'PrimaryZip_Perm':'Address_zip','PrimaryCountry_Perm':'Address_country','PrimaryCity_Perm':'Address_city','PrimaryState_Perm':'Address_state','PrimaryStreet1_Perm':'Address_address'})
#------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#HERE WE MODIFY SOME COLUMNS TO MAKE IT NORMALISED
#To make it visually beautifull is to Capitalize the First Character of Each Word in selected rows:
df_trans['First_name'] = df_trans['First_name'].str.capitalize()
df_trans['Middle_name'] = df_trans['Middle_name'].str.capitalize()
df_trans['Last_name'] = df_trans['Last_name'].str.capitalize()
df_trans['Company_address'] = df_trans['Company_address'].str.title()
df_trans['Address_address'] = df_trans['Address_address'].str.title()

#We need to delete all the ENDs of Addresses like St. , Blvd, Rd  or Dr to Normalise them:
df_trans['Address_address'] = df_trans['Address_address'].str.replace(r'st.$', '', regex=True).str.strip()
df_trans['Address_address'] = df_trans['Address_address'].str.replace(r'blvd$', '', regex=True).str.strip()
df_trans['Address_address'] = df_trans['Address_address'].str.replace(r'rd$', '', regex=True).str.strip()
df_trans['Address_address'] = df_trans['Address_address'].str.replace(r'dr$', '', regex=True).str.strip()
df_trans['Address_address'] = df_trans['Address_address'].str.replace(r'st$', '', regex=True).str.strip()
df_trans['Company_address'] = df_trans['Company_address'].str.replace(r'st.$', '', regex=True).str.strip()
df_trans['Company_address'] = df_trans['Company_address'].str.replace(r'rd$', '', regex=True).str.strip()
df_trans['Company_address'] = df_trans['Company_address'].str.replace(r'st$', '', regex=True).str.strip()

#In Date_register we dont need the Hour:Minute:Second so delete them easily:
df_trans['Date_register'] = df_trans['Date_register'].str[:-8]
#-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Create the columns missed:
#Spliting the full address we get: Address_street & Address_street_number 
df_trans['Address_street_number']=df_trans['Address_address'].str.extract(r'(^\d{,5})')
df_trans['Address_street']=df_trans['Address_address'].str.extract(r'(\D{,15}$)')
df_trans['Address_address'] = df_trans['Address_street'].fillna('').map(str) + " " + df_trans['Address_street_number'].fillna('').map(str)

#And same for Company_address
df_trans['Company_address_street_number']=df_trans['Company_address'].str.extract(r'(^\d{,5})')
df_trans['Company_address_street']=df_trans['Company_address'].str.extract(r'(\D{,15}$)')
df_trans['Company_address'] = df_trans['Company_address_street'].fillna('').map(str) + " " + df_trans['Company_address_street_number'].fillna('').map(str)

#Concatenating we get Full_name:
df_trans['Full_name'] = df_trans['First_name'].map(str) + " " + df_trans['Middle_name'].fillna('').map(str) + " " + df_trans['Last_name'].map(str)
#-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------

Selected_df = df_trans[['Email', 'Username', 'Phone', 'First_name', 'Middle_name', 'Last_name', 'Full_name' , 'Gender', 'Marital_status', 'Date_of_birth', 'Date_register', 'Company', 'Company_address' ,'Professional_role', 'Address_address', 'Address_city', 'Address_state', 'Address_street', 'Address_street_number', 'Address_zip', 'Address_country']]
Selected_df
```
