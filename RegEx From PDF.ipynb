```python
!pip install pdfx &> /dev/null
import pdfx
pdf = pdfx.PDFx("https://www.boe.es/boe/dias/2015/05/27/pdfs/BOE-A-2015-5834.pdf")
text = pdf.get_text()
print(text)
```

```python
#DNI always have 8 Digits and 1 letter, And NIE starts with 1 letter, 7 digits and ends 1 letter
import re
DNI_list = re.findall("[A-Z]{,1}[0-9]{7,8}[A-Z]{,1}", text,)
DNI_list
```
