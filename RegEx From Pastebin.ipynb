```python
import requests # requests module of python

url = 'https://pastebin.com/q9bn7Nby' # url of paste

def rreplace(s, old, new): #To get the Url in RAW mode we replace the LAST / by /raw/
    return (s[::-1].replace(old[::-1],new[::-1], 1))[::-1]

raw_url = rreplace(url,'/','/raw/')

r = requests.get(raw_url) # response will be stored from url
email_list = r.text.split()  # Create a list whith the raw text from url
print(email_list) # prints content
```

```python
import re

pattern = re.compile(r'[\w.]+@gmail.com') #This will select any value thats ENDS with @gmail.com in the list created
filtered_email = list(filter(pattern.findall, email_list))
print(filtered_email)
```
