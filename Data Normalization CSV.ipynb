```python
!pip install --upgrade deepl
!pip install --upgrade pandas
import pandas as pd
import io
import requests # requests module of python
import deepl
import re
translator = deepl.Translator("0995d402-5b66-68f0-d6dc-68c7622c351a:fx")
```

```python
url = 'https://pastebin.com/MmryYxH7' # url of paste

def rreplace(s, old, new): #To get the Url in RAW mode we replace the LAST / by /raw/
    return (s[::-1].replace(old[::-1],new[::-1], 1))[::-1]

raw_url = rreplace(url,'/','/raw/')
text = requests.get(raw_url).text # response will be stored as a TEXT
print(text) # prints content
```

```python
string = text.replace("Умное голосование 10k Mosсow RUSIA", "") #We remove the first line that is the name of the file
string = string.replace("\r\n", " ") # We delete all the Carriage Return+Line Feed
# matches all email for the next step
p = re.compile('[\w.-]+@\w+\.[a-z]{,3}')
# We introduce the backslash after every email to determine the lines.
new_string = p.sub(r'\n\g<0>', string) 
print(new_string)
```

```python
# Translate text into a target language, in this case, English
translated = translator.translate_text(new_string, target_lang="EN-GB")
translated = translated.text
print(translated)
```

```python
Lines_list = translated.split('\n') # Create a list F
Inter = [sub.replace(';', '_', 3) for sub in Lines_list] 
Inter = [sub[::-1].replace(';', '_', 1)[::-1] for sub in Inter] #Just reverse the string, replace first occurrence and reverse it again
Inter = [sub.replace(';', '') for sub in Inter]
Inter = [sub.replace('_', ';') for sub in Inter]
translated_text = '\n'.join(Inter)
translated_text.encode('utf8')
print(translated_text)
```

```python
data = io.StringIO(translated_text)
df = pd.read_csv(data, sep=";" , index_col=None)
pd.set_option('display.max_colwidth', -1)
df
```

```python
#Fields not found to normalise: Username, Phone, password_plain, password_hash, Fax , Secondary_last_name, Marital_status , date_miscellaneous , Iban, Company , Company_address , Professional_role , web_address, Address_address , Address_city , Address_state, Address_street ,Address_street_number ,Address_zip, address_country , ip_ipv4, ip_ipv6, avatar
# Date_update , Date_miscellaneous ,Iban, Credit_card_number, Credit_card_last4  Credit_card_exp_date, credit_card_exp_month, Credit_card_exp_year, Credit_card_type , web_address, ip_ipv4 , ip_ipv6 , token , avatar# 
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

#Rename Fields
dfrenamed = df.rename(columns={' POST':'Email','NAME':'Full_name','DD':'Date_of_birth','Employer':'Company','Address':'Full_Address'})

#Split Fields to create new fields: Full name, First_name , Middle_name , Last_name , Address_street , Address_street_number.
dfrenamed['Middle_name'], dfrenamed['First_name'], dfrenamed['Last_name'] = dfrenamed['Full_name'].str.split(' ', 2).str
dfrenamed['Address_street'], dfrenamed['Address_street_number'] = dfrenamed['Full_Address'].str.split(',', 1).str


# NORMALISATION
dfrenamed['Date_of_birth'] = dfrenamed['Date_of_birth'].str.replace(r'.', '-', regex=True).str.strip()
dfrenamed['Company'] = dfrenamed['Company'].str.replace(r'JOINT-STOCK COMPANY', ' JSC.:', regex=True).str.strip()
dfrenamed['Company'] = dfrenamed['Company'].str.replace(r'JOINT STOCK COMPANY', ' JSC.:', regex=True).str.strip()
dfrenamed['Company'] = dfrenamed['Company'].str.replace(r'LIMITED LIABILITY COMPANY', ' LLC.:', regex=True).str.strip()
dfrenamed['Company'] = dfrenamed['Company'].str.replace(r'""', '\'', regex=True).str.strip()
dfrenamed['Company'] = dfrenamed['Company'].str.replace(r'"', '', regex=True).str.strip()
dfrenamed['Address_street_number'] = dfrenamed['Address_street_number'].str.replace(r'SQ', 'Flat Nº:', regex=True).str.strip()





Selected_df = dfrenamed[['Email','First_name', 'Middle_name' ,'Last_name', 'Full_name' , 'Date_of_birth' ,'Company','Address_street','Address_street_number']]
Selected_df
```

```python
#Transform to Json format.
js = Selected_df.to_json(orient = 'records', force_ascii=False, lines=True)
print(js)
```

```python
import zlib
import zipfile

def compress(file_names):
    print("File Paths:")
    print(file_names)

    path = "C:/data/"

    # Select the compression mode ZIP_DEFLATED for compression
    # or zipfile.ZIP_STORED to just store the file
    compression = zipfile.ZIP_DEFLATED

    # create the zip file first parameter path/name, second mode
    zf = zipfile.ZipFile("RAWs.zip", mode="w")
    try:
        for file_name in file_names:
            # Add file to the zip file
            # first parameter file to zip, second filename in zip
            zf.write(path + file_name, file_name, compress_type=compression)

    except FileNotFoundError:
        print("An error occurred")
    finally:
        # Don't forget to close the file!
        zf.close()


file_names= ["test_file.txt", "test_file2.txt"]
compress(file_names)
```
