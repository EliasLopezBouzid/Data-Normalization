```python
!pip install --upgrade pandas
import pandas as pd

import requests # requests module of python
import re

from pandas.io.json import json_normalize
import numpy as np
```

```python
url = 'https://pastebin.com/sGynD3SF' # url of paste

def rreplace(s, old, new): #To get the Url in RAW mode we replace the LAST / by /raw/
    return (s[::-1].replace(old[::-1],new[::-1], 1))[::-1]

raw_url = rreplace(url,'/','/raw/')
text = requests.get(raw_url).text # response will be stored as a TEXT
print(text) # prints content
```

```python
string = text.replace("\r\n", "") # We delete all the backslash
# matches all the starts of lines for the next step
p = re.compile('{"_id":{')
# We introduce the line jump after every {"_id":{ to determine the lines.
string = p.sub(r'\n{"_id":{', string) 
newstring = string.replace(" ", "") # We delete all the spaces
print(newstring)
```

```python
#Finally we create the Dataframe thanks to the method pd.read_json
df = pd.read_json(newstring, lines=True)
def unpack(x):
    rv = []
    for v in x:
        if isinstance(v, dict):
            rv.append([*v.values()][0])
        else:
            rv.append(v)
    return rv
df = df.apply(unpack)
pd.set_option('display.max_columns', None) #This line makes to shows all the columns.
pd.set_option('display.max_colwidth', -1)
df
```

```python
# Now we will delete all fields that have a percentage of NaN > 80% in order to see better and choose the desired fields to normalize.
df_clear = df.dropna(axis=1, thresh=int(0.2*len(df)))
df_clear
```

```python
#Fields to normalize:

#Fields not found to normalise: password_hash, Fax , Middle_name, Secondary_last_name, Marital_status , date_miscellaneous , Iban, Company , Company_address , Professional_role , web_address, Address_address , Address_city , Address_state, Address_street ,Address_street_number ,Address_zip, address_country , ip_ipv4, ip_ipv6, avatar
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Renaming columns to match the selected normalisation table:
df_trans = df_clear.rename(columns={'email':'Email', 'login':'Username', 'password':'Password_plain','phone':'Phone','first_name':'First_name','last_name':'Last_name','gender':'Gender'})
df_trans = df_trans.rename(columns={'created':'Date_register','birth':'Date_of_birth','modified':'date_update','card_id':'Credit_card_number', 'payments':'Credit_card_number2','stripe_cc_token':'Token'})
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#NORMALISATION
#In Date_register & Date_of_birth we dont need the Hour:Minute:Second so delete them easily:
df_trans['Date_of_birth'] = df_trans['Date_of_birth'].str[:10]
df_trans['Date_register'] = df_trans['Date_register'].str[:10]

#Credit_card_number are shown as Float so let change to Int
df_trans['Credit_card_number'] = df_trans['Credit_card_number'].astype("Int64")

#In gender Change F to females and M to male:
df_trans['Gender'] = df_trans['Gender'].str.replace('M$', 'MALE')
df_trans['Gender'] = df_trans['Gender'].str.replace('F$', 'FEMALE')

#In the Phone field lets take the las 10 digits in order to have a clean view.
df_trans['Phone'] = df_trans['Phone'].str.extract(r'(\d{10})$')

#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#Fields to Create: Full name, Credit_card_last4 , Credit_card_exp_date , credit_card_exp_month , Credit_card_exp_year , Credit_card_type.

#Concatenating we get Full_name:
df_trans['Full_name'] = df_trans['First_name'].map(str) + " " + df_trans['Last_name'].map(str)

#We create Credit_card_last4 from Credit_card_number
df_trans['Credit_card_number'] = df_trans['Credit_card_number'].astype(str)
df_trans['Credit_card_last4'] = df_trans['Credit_card_number'].str.extract(r'(\d{4})$')


#Here we have a high dificult step parsing the column Credit_card_number2 in order to extract info from it:
Auxdf = pd.json_normalize(df_trans['Credit_card_number2']) #Inside the column we have another Json so we extract it into Auxdf
df_trans['Credit_card_number2'] = Auxdf['cards'] #We take the column cards from Auxdf
df_trans['Credit_card_number2'] = df_trans['Credit_card_number2'].astype(str) #This column now is a Dictionary so we transform into Stringç

#We have transformed the column Credit_card_number2 so we can extract information from it with Regular Expresions and create the new Fields:
df_trans['Credit_card_type']=df_trans['Credit_card_number2'].str.extract(r'\'brand\': \'(\w+)\',')
df_trans['Credit_card_last4-2']=df_trans['Credit_card_number2'].str.extract(r'\'last4\': \'(\d+)\',')
df_trans['credit_card_exp_month']=df_trans['Credit_card_number2'].str.extract(r'\'exp_month\': (\d+),')
df_trans['Credit_card_exp_year']=df_trans['Credit_card_number2'].str.extract(r'\'exp_year\': (\d+),')

#The field Credit_card_last4 dont includes the data extracted from Credit_card_number2 IN Credit_card_last42 so lets combine:
df_trans['Credit_card_last4'] = df_trans['Credit_card_last4'].combine_first(df_trans['Credit_card_last4-2'])

#We delete from the field token the starts " tok_ "
df_trans['Token'] = df_trans['Token'].str[4:]

#In the process we made some conversions and NAN turned into <NA> so revert it
df_trans = df_trans.replace('None', np.NaN)
df_trans = df_trans.replace('<NA>', np.NaN)

Selected_df = df_trans[['Email', 'Username', 'Password_plain' ,'Phone', 'First_name','Last_name', 'Full_name' ,'Gender', 'Date_of_birth' ,'Date_register', 'Credit_card_number' ,'Credit_card_last4','credit_card_exp_month','Credit_card_exp_year','Credit_card_type', 'Token']]
Selected_df
```

```python
js = Selected_df.to_json(orient = 'records', force_ascii=False, lines=True)
print(js)
```
